```python
import re
```

```python
def decode_line(encoded_str:str, spell_num=False):
    if spell_num:
        spell_num = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
        str_num = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
        num_regex = '|'.join(["(?:{})".format(num) for num in spell_num])
    
        match_el = re.findall(num_regex, encoded_str)
        if len(match_el) > 0:
            match_num = match_el[0]
            spell_index = spell_num.index(match_num)
            encoded_str = encoded_str.replace(match_num, str_num[spell_index])
        if len(match_el) > 1:
            match_num = match_el[-1]
            spell_index = spell_num.index(match_num)
            encoded_str = encoded_str.replace(match_num, str_num[spell_index])

    encoded_str = re.sub("\D", "", encoded_str)
    

    return encoded_str
    
```

```python
def decode_gen(encoded_str:str, spell_num=False)->int:

    encoded_lines = encoded_str.splitlines()

    i=0
    while i < len(encoded_lines):
        decoded_line = decode_line(encoded_lines[i], spell_num)
        if decoded_line.isnumeric():
            two_digit_num = decoded_line[0]+ decoded_line[-1] # if len(decoded_line) > 1 else '')
            print(encoded_lines[i], '->', two_digit_num)
            yield int(two_digit_num)
        else:
            yield 0
        i+=1
    return None
    
```

```python
with open('./input', 'r') as f:
    encoded_str = f.read()
```

```python
total=0
for num in decode_gen(encoded_str, spell_num=True):
    total+=num
```
